```python
"""
Task 3: API Fetcher & JSON Parser (Real-World Data Handling)
-------------------------------------------------------------
This script fetches live weather data using the Open-Meteo API
and displays summarized insights in a terminal dashboard format.
"""

import requests
import json
from datetime import datetime


def fetch_weather(latitude=31.5497, longitude=74.3436):
    """Fetches weather data for a given latitude and longitude."""
    url = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current_weather=true"

    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        return response.json()

    except requests.exceptions.RequestException as e:
        print(f"âŒ Error fetching data: {e}")
        return None


def display_dashboard(data):
    """Displays weather info in a clean terminal dashboard format."""
    city = "Lahore, Pakistan"
    current = data["current_weather"]

    # Extract key information
    temp = current["temperature"]
    wind = current["windspeed"]
    code = current["weathercode"]
    time_utc = current["time"]

    # Convert time to readable format
    try:
        time_local = datetime.strptime(time_utc, "%Y-%m-%dT%H:%M").strftime("%Y-%m-%d %H:%M")
    except:
        time_local = time_utc

    # Dashboard layout
    print("=" * 37)
    print("ğŸŒ¤ï¸  WEATHER DASHBOARD -", city.upper())
    print("=" * 37)
    print(f"ğŸ“… Date & Time : {time_local} UTC")
    print(f"ğŸŒ¡ï¸ Temperature : {temp}Â°C")
    print(f"ğŸ’¨ Wind Speed  : {wind} km/h")
    print(f"ğŸŒ¦ï¸ Weather Code: {code}")
    print("-" * 37)
    print("âœ… Data Source : Open-Meteo API")
    print("âœ… Status      : Live Data Fetched Successfully")
    print("=" * 37)


def save_to_json(data):
    """Saves the API response to a local JSON file."""
    with open("weather_dashboard.json", "w") as f:
        json.dump(data, f, indent=4)
    print("\nğŸ’¾ Data saved to 'weather_dashboard.json'")


def main():
    print("  API Fetcher & JSON Parser\n")
    data = fetch_weather()

    if data:
        display_dashboard(data)
        save_to_json(data)
    else:
        print("\nâš ï¸ Could not retrieve weather data.")


# Run main program
if __name__ == "__main__":
    main()
```
